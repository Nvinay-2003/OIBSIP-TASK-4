# Importing Necessary Packages

```python
import numpy as np
import pandas as pd
import matplotlib as plt
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score
```

# Data Collection and Pre-Processing

```python
df = pd.read_csv("spam.csv", encoding='<latin-1>')
df=df.drop(columns=["Unnamed: 2","Unnamed: 3","Unnamed: 4"])
print(df)
```

```python
df.describe()
```

```python
#removing rows with missing values
df = df.dropna()
```

```python
df.describe()
```

```python
df.shape
```

# Using LabelEncoder to encode the target column

```python
#Making target and feature column
X=df['v2']
y=df['v1']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

Ham-0
Spam-1

```python
print(X)
```

# Training,Testing and Splitting 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

# Feature Extraction

```python
# transform the text data to feature vectors
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase='True')

X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)

# convert y_train and y_test values as integers
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(X_train_features)
```

# Building the Model

```python
# Import the necessary libraries
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Build the Logistic Regression model
lr_classifier = LogisticRegression()
lr_classifier.fit(X_train_features, y_train)
lr_predictions = lr_classifier.predict(X_test_features)
lr_accuracy = accuracy_score(y_test, lr_predictions)
print("Logistic Regression accuracy:", lr_accuracy)

# Build the Naive Bayes model
nb_classifier = MultinomialNB()
nb_classifier.fit(X_train_features, y_train)
nb_predictions = nb_classifier.predict(X_test_features)
nb_accuracy = accuracy_score(y_test, nb_predictions)
print("Naive Bayes accuracy:", nb_accuracy)

# Build the SVM model
svm_classifier = SVC(kernel='linear')
svm_classifier.fit(X_train_features, y_train)
svm_predictions = svm_classifier.predict(X_test_features)
svm_accuracy = accuracy_score(y_test, svm_predictions)
print("SVM accuracy:", svm_accuracy)

# Build the Random Forest model
rf_classifier = RandomForestClassifier(n_estimators=100)
rf_classifier.fit(X_train_features, y_train)
rf_predictions = rf_classifier.predict(X_test_features)
rf_accuracy = accuracy_score(y_test, rf_predictions)
print("Random Forest accuracy:", rf_accuracy)
```

I have selected the svm model to predict as it has more accuracy while compared to other models.

# New Predictions
